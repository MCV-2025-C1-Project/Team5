```python
img_rgb
```

```python

r = img_rgb[:, :, 0]
g = img_rgb[:, :, 1]
b = img_rgb[:, :, 2]
# Concatenate in CbCr order (replace the YCrCb that OpenCV returns)
img_rg = np.stack([r, g], axis=-1)
img_rb = np.stack([r, b], axis=-1)
img_rg_reshape = img_rg.reshape(
    (img_rg.shape[0]*img_rg.shape[1], img_rg.shape[2]))
img_rb_reshape = img_rb.reshape(
    (img_rb.shape[0]*img_rb.shape[1], img_rb.shape[2]))
```

```python
img_rg_reshape
```

```python
plt.imshow(img_rgb)
plt.show()
```

```python
from src.descriptors.histogram import compute_histogram
hist_rb, bin_edges_rg = compute_histogram(
   img_rb_reshape, values_per_bin=1, dimension=2, density=False)
```

```python
hist_rg
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Construct arrays for the anchor positions of the 16 bars.
xpos, ypos = np.meshgrid(bin_edges_rg[0][:-1] + 0.25, bin_edges_rg[1][:-1] + 0.25, indexing="ij")
xpos = xpos.ravel()
ypos = ypos.ravel()
zpos = 0

# Construct arrays with the dimensions for the 16 bars.
dx = dy = 0.5 * np.ones_like(zpos)
dz = hist_rb.ravel()

ax.bar3d(xpos, ypos, zpos, dx, dy, dz, zsort='average')

plt.show()
```

```python
# Plot heatmap
plt.figure(figsize=(7, 6))
plt.imshow(
    hist_rb.T,                # transpose to match axes correctly
    origin='lower',        # puts the lowest values at bottom-left
    extent=[bin_edges_rg[0][0], bin_edges_rg[0][-1], bin_edges_rg[1][0], bin_edges_rg[1][-1]],
    cmap='viridis',        # colormap
    aspect='auto'
)

plt.colorbar(label='Frequency')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Histogram Heatmap')
plt.show()
```
